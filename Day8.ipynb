```python
#load the datset
import pandas as pd
import numpy as np
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
```

```python
test_df.head()
```

```python
#concatenate the train and test dataframes
full_df = pd.concat([train_df, test_df], ignore_index=True)
```

```python
full_df.head()
```

```python
#drop unecessary columns
full_df.drop(columns=["size_units", "lot_size_units", "lot_size","zip_code"], inplace=True)
full_df.head()
```

```python
full_df.info()
```

```python
#change the data type of bath column
full_df['baths'] = full_df['baths'].astype(int)
full_df.info()
```

```python
full_df.isna().sum()
```

```python
#identify the input and output
X = full_df.drop('price',axis=1)
y = full_df['price']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
house_price_mlr_model = LinearRegression()
house_price_mlr_model.fit(X_train, y_train)
```

```python
house_price_mlr_model.coef_
```

```python
house_price_mlr_model.intercept_
```

```python
#Test the model
y_pred = house_price_mlr_model.predict(X_test)
y_pred[:5]
```

```python
y_test[:5]
```

```python
#evaluate the model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)   
rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)   
print(mae,rmse,r2)
```

```python
#training the model on scaled or normalized data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled[:5]
```

```python
from sklearn.linear_model import LinearRegression
house_price_mlr_model_new = LinearRegression()
house_price_mlr_model_new.fit(X_train_scaled, y_train)
```

```python
house_price_mlr_model_new.intercept_
```

```python
house_price_mlr_model_new.coef_
```

```python
#predictions
y_pred_new = house_price_mlr_model_new.predict(X_test_scaled)
y_pred_new[:5]
```

```python
#evaluate the model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, root_mean_squared_error
mae = mean_absolute_error(y_test, y_pred_new)   
rmse = root_mean_squared_error(y_test, y_pred_new)
r2 = r2_score(y_test, y_pred_new)
print(mae,rmse,r2)
```

```python
#save the model
import joblib as jb
jb.dump(house_price_mlr_model_new, "House_Price_Predictor.pkl")
```
