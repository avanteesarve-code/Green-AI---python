```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('Bitcoin_Price_Dataset_2014_2023.csv')
print("Original Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
print("Missing Values:")
print((df.isnull().sum()/len(df)*100).round().astype(int))
```

```python
df.fillna({
    'Daily_Return': df['Daily_Return'].median()}, inplace=True)
```

```python
df.fillna({'MA_7': df['MA_7'].median()}, inplace=True)
```

```python
type(df['Volume'].mode()[0])
```

```python
#Drop column with too many missing values
df.drop(columns=['MA_30'], inplace=True)   

print("missing values after imputation:")
print(df.isnull().sum())
```

```python
Q1 = df['Close'].quantile(0.25)
Q3 = df['Close'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)
```

```python
df['Return_Type'] = df['Daily_Return'].apply(
    lambda x: 'Positive' if x > 0 else 'Negative'
)
```

```python
plt.figure()
sns.countplot(x='Return_Type', data=df)
plt.title("Count of Positive vs Negative Days")
plt.show()
```

```python
plt.figure()
sns.histplot(df['Daily_Return'], bins=30, kde=True)
plt.title("Daily Return Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Correlation Heatmap - Bitcoin Dataset")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Year', y='Close', data=df)

plt.title("Close Price Distribution by Year")
plt.xticks(rotation=45)
plt.show()
```
