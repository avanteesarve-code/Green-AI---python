```python
import seaborn as sns
data = sns.load_dataset('titanic')
data.head()
```

```python
data.info()
```

```python
data.sample(10)
```

```python
#Deleting the single column
data.drop("embarked",axis = 1, inplace = True)
```

```python
data.head()
```

```python
#deleting the single column
data.drop("class", axis = 1, inplace = True)
data.head()
```

```python
data.isna().sum()
```

```python
data['age'].fillna(data['age'].mean(), inplace = True)
data.isna().sum()
```

```python
data['age'] = data['sex'].astype(str)
data.info()
```

```python
#One hot encoding to encode the 'sex' column
import pandas as pd
data_encoded = pd.get_dummies(columns = ['sex'], data = data, dtype = int, drop_first = True)
data_encoded.head()
```

```python
#save the file
data_encoded.to_csv('titanic_encoded.csv', index = False)
```
