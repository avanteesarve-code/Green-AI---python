```python
limit = 8000
transactions = []

for i in range(5):
    amount = int(input("Enter transaction amount: "))
    transactions.append(amount)

total = 0

for i in range(len(transactions)):
    total = total + transactions[i]
    
    if total >= limit:
        print("Limit reached after", i+1, "transactions")
        
        if total > limit:
            print("Balance is over by", total - limit)
        break
else:
    print("Limit not reached after all transactions")
```
